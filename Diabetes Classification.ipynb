```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Diabetes_XTrain.csv")
df2 = pd.read_csv("Diabetes_YTrain.csv")
df3 = pd.read_csv("Diabetes_Xtest.csv")

print(df.shape)
print(df2.shape)
print(df3.shape)
```

```python
print(df.columns)
print(df2.columns)
```

```python
data = df.values
data2 = df2.values
data3 = df3.values
```

```python
X = data[:,1:]
Y = data2[:,0]
Z = data3[:,1:]

print(X.shape,Y.shape,Z.shape)
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(X,Y,queryPoint,k=5):
    
    vals = []
    m = X.shape[0]
    
    for i in range(m):
        d = dist(queryPoint,X[i])
        vals.append((d,Y[i]))
        
    
    vals = sorted(vals)
    # Nearest/First K points
    vals = vals[:k]
    
    vals = np.array(vals)
    
    #print(vals)
    
    new_vals = np.unique(vals[:,1],return_counts=True)
    #print(new_vals)
    
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    
    return pred
```

```python
outcome = []
for i in range(192):
    pred = knn(X,Y,Z[i])
    ans = int(pred)
    outcome.append(ans)
    print(ans)
print (outcome)
```

```python
dfn = pd.DataFrame()
dfn['Outcome'] = outcome
dfn.to_csv("submission.csv", sep=',',index=False)
```
